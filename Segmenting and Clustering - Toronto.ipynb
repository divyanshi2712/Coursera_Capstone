<h1> Part 1

```python
import numpy as np 
import pandas as pd 
```

```python
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
tables = pd.read_html(link,header=0)[0]
tables
```

<h2><i><u>Removing rows with Not Assigned Borough

```python
tables.columns=['PostalCode','Borough','Neighborhood']
table_ext = tables[tables['Borough'] != 'Not assigned']
table_ext
```

<h2><i><u>Combining Neighborhood with same PostalCode</h1>

```python
# Group by postcode
table_grp = table_ext.groupby('PostalCode')
column_names = ['PostalCode', 'Borough', 'Neighborhood']
df_neigh = pd.DataFrame(columns=column_names)

def combine_neighbor(grp):
    return table_grp.get_group(grp)['Borough'].iloc[0], ", ".join(table_grp.get_group(grp)['Neighborhood'])
```

```python
for grp in table_grp.groups.keys():
    borough, neighbor = combine_neighbor(grp)
    df_neigh = df_neigh.append({'PostalCode' : grp,
                                          'Borough': borough,
                                          'Neighborhood': neighbor
                                         }, ignore_index=True)
df_neigh
```

<h2><i><u> Changing value For Not Assigned Neighbor

```python
print("Checking number of rows with Not Assigned Neighbor BEFORE processing...") 
print(df_neigh[df_neigh['Neighborhood'].str.contains('Not assigned')])
```

```python
df_neigh.loc[df_neigh['Neighborhood'] == 'Not assigned', 'Neighborhood']= df_neigh[df_neigh['Neighborhood'] == 'Not assigned']['Borough']

print("Checking number of rows with Not Assigned Neighbor AFTER processing...") 
print(df_neigh[df_neigh['Neighborhood'].str.contains('Not assigned')])
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
LIENT_ID = 'B0VDENVC0SXIGVA3YGD0MEKKB4FBQYFCHT2VAL3QAU1KY5DD'
CLIENT_SECRET = 'EHREYC4XEDRHGVCPBZUQX3W4YULHK0RZLMKNOVE2HI4TR0OM' # your Foursquare Secret
VERSION = '20190929' # Foursquare API version
radius = 500
LIMIT = 200

def getVenues(names, latitude, longitude):
    
    list_venues=[]
    for name, lat, lng in zip(names, latitude, longitude):
            
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        results = requests.get(url).json()["response"]['groups'][0]['items']
       
        list_venues.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in list_venues for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
df_toronto = getVenues(names=df_neigh_toronto['Neighborhood'],
                                   latitude=df_neigh_toronto['Latitude'],
                                   longitude=df_neigh_toronto['Longitude'],
                                  )
df_toronto.head()
```

<h2><i><u> Analyzing each Neighborhood

```python
# Convert each category to one hot
toronto_onehot = pd.get_dummies(df_toronto[['Venue Category']], prefix="", prefix_sep="")
# Add neighborhood back
toronto_onehot['Neighborhood'] = df_toronto['Neighborhood'] 
# move neighborhood column to the first column
toronto_onehot = toronto_onehot.set_index('Neighborhood').reset_index()

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_grouped.shape
```

<h2><i><u> Printing each Neighborhood along with 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

<h2><i><u> Putting it into pandas dataframe

```python
# funtions for Sorting in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# create new dataframe and display the top 10 venues for each neighborhood.
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

<h1><i><u> Clustering

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merge = df_neigh_toronto

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merge = toronto_merge.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merge.head()
```

<h3 color="blue"> Creating Map

```python
# creating map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# setting color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# adding markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merge['Latitude'], toronto_merge['Longitude'], toronto_merge['Neighborhood'], toronto_merge['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
